<h1>‚óè	Plot multiple lines on the same graph.</h1>

```python
import matplotlib.pyplot as plt


X=[1,2,3,4,5,6]
Y1=[2,4,6,8,10,12]
Y2=[1,2,3,4,5,6]
Y3=[1/2,2/2,3/2,4/2,5/2,6/2]

plt.plot(X,Y1, label = 'line1:y1=2x')
plt.plot(X,Y2, label = 'line2:y2=x')
plt.plot(X,Y3, label = 'line3:y3=x/2')

plt.xlabel("x-axis")

plt.ylabel("y-axis")

plt.title("simple line plot graph")
plt.legend()

plt.show()


```

<h1>‚óè	Add markers to a line plot</h1>

```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[1,8,27,48,]

plt.plot(x,y,marker='o',label= "label: y=x**3")

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

<h1>‚óè	Customize line styles and colors.</h1>

```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[1,8,27,48,]

plt.plot(x,y,marker='o',label= "label: y=x**3",linestyle="--",color="brown")

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

<h1>2. Scatter Plots:</h1>

<h1>‚óè	Create a basic scatter plot.</h1>

```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[2,4,6,8]

plt.scatter(x,y)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

<h1>‚óè	Customize marker styles, sizes, and colors.</h1>

```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[2,4,6,8]

plt.scatter(x,y,marker='o',label='simple line', linestyle='--', color='brown', s=100) #s is size ok

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

<h1>‚óè	Plot multiple scatter plots on the same graph.</h1>

```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y1=[2,4,6,8]
y2=[4,8,12,16]
y3=[0,1,2,3]

plt.scatter(x,y1)
plt.scatter(x,y2)
plt.scatter(x,y3)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```


<h1>‚óè	Add labels and titles to the scatter plot.</h1>

```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y1=[2,4,6,8]
y2=[4,8,12,16]
y3=[0,1,2,3]

plt.scatter(x,y1,label='label:y=x')
plt.scatter(x,y2,label='label:y=2x')
plt.scatter(x,y3,label='label:y=x-1')
plt.title('this my ful feature scatter graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.legend()
plt.show()
```

<h1>3.	Bar Charts:</h1>
<h1>vertical bar chart</h1>

```python
import matplotlib.pyplot as plt
# import seaborn as sns

a=[2,4,6,8]
b=[1,3,5,7]

plt.bar(a,b)
# sns.barplot(x=a,y=b,color='black')
plt.title('barchart with matplotlib')
plt.xlabel(a)
plt.ylabel(b)
plt.show()
```

<h1>‚óè	Create a horizontal bar chart.</h1>


```python
import matplotlib.pyplot as plt
# import seaborn as sns

a=[2,4,6,8]
b=[1,3,5,7]

plt.barh(a,b)
# sns.barplot(x=a,y=b,color='black')
plt.title('barchart with matplotlib')
plt.xlabel(a)
plt.ylabel(b)
plt.show()
```

<h1>‚óè	Customize bar colors and add edge colors.</h1>

<h1>4.	Pie Charts:</h1>

```python
import matplotlib.pyplot as plt
a=[2,4,6,8]
b=[1,3,5,7]

plt.pie(a,b, autopct='%1.1f%%', startangle=90)
plt.title('my pie chart for basic understanding')
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]  # Values for each section
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']  # Custom colors

# Create the pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

# Equal aspect ratio ensures that pie chart is drawn as a circle.
plt.axis('equal')

# Add a title
plt.title('Basic Pie Chart')

# Show the chart
plt.show()
```

<h1>‚óè	Customize slice colors and add a legend.</h1>

```python
import matplotlib.pyplot as plt
a = [2, 4, 6, 8]
labels = ['Section 1', 'Section 2', 'Section 3', 'Section 4']
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']  # Light red, blue, green, orange
plt.pie(a, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Customized Pie Chart with Colors and Legend')
plt.axis('equal')
plt.legend(labels, title="Sections", loc="best")
plt.show()
```

<h1>‚óè	Explode a slice of the pie chart.</h1>

```python
import matplotlib.pyplot as plt
a = [2, 4, 6, 8]
labels = ['Section 1', 'Section 2', 'Section 3', 'Section 4']
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
explode = (0, 0, 0.1, 0)
plt.pie(a, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, explode=explode)
plt.title('Pie Chart with Exploded Slice')
plt.axis('equal')
plt.legend(labels, title="Sections", loc="best")
plt.show()
```
