<h1>1.	Merge two DataFrames on a single key.</h1>

```python
import pandas as pd

data1 = {'r_no':[1,2,3],
          'name':['ahsan','mohsin','abdul_rehman']}
df1 = pd.DataFrame(data1)

data2 = {'r_no':[1,6,3],
          'name':['hakeem','shamy','nabi']}
df2 = pd.DataFrame(data2)

msk = pd.merge(df1,df2, on='r_no')

msk
```

<h1>2.	Merge two DataFrames on multiple keys.</h1>

```python
import pandas as pd

data1 = {'r_no':[1,2,3],
          'name':['ahsan','mohsin','abdul_rehman']}
df1 = pd.DataFrame(data1)

data2 = {'r_no':[1,6,3],
          'name':['ahsan','shamy','abdul_rehman']}
df2 = pd.DataFrame(data2)

msk = pd.merge(df1,df2, on=['r_no','name'])

msk
```

<h1>4.	Concatenate two DataFrames along rows.</h1>

```python
import pandas as pd

data1 = {'r_no':[1,2,3],
          'name':['ahsan','mohsin','abdul_rehman']}
df1 = pd.DataFrame(data1)

data2 = {'r_no':[1,6,3],
          'name':['ahsan','shamy','abdul_rehman']}
df2 = pd.DataFrame(data2)

con = pd.concat([df1,df2])
con
```

<h1>5.	Concatenate two DataFrames along columns.</h1>

```python
import pandas as pd

data1 = {'r_no':[1,2,3],
          'name':['ahsan','mohsin','abdul_rehman']}
df1 = pd.DataFrame(data1)

data2 = {'r_no':[1,6,3],
          'name':['ahsan','shamy','abdul_rehman']}
df2 = pd.DataFrame(data2)

con = pd.concat([df1,df2],axis=1)
con
```

<h1>6.	Concatenate a list of DataFrames.</h1>

<h1>7.	Reshape data using the melt function to go from wide to long format.</h1>

```python
import pandas as pd
data = {'name':['ahsan','mohsin','ali'],
        'urdu':[10,16,20],
         'sst':[15,14,16]}
df = pd.DataFrame(data)


rshap = pd.melt(df,id_vars=['name'],var_name='subject',value_name='score')
rshap
```

<h1>8.	Create a pivot table to summarize data</h1>

```python
import pandas as pd
data = {'name':['ahsan','mohsin','ali'],
        'urdu':[10,16,20],
         'sst':[15,14,16]}
df = pd.DataFrame(data)
rshap = pd.melt(df,id_vars=['name'],var_name='subject',value_name='score')
spt = pd.pivot_table(rshap, index='name', columns='subject', values='score', aggfunc='mean')
spt
```

<h1>9.	Group data by one or more columns and perform aggregation functions (e.g., sum, mean, count).</h1>

```python
import pandas as pd
data = {'name':['ahsan','mohsin','ali'],
        'urdu':[10,16,20],
         'sst':[15,14,16]}
df = pd.DataFrame(data)
rshap = pd.melt(df,id_vars=['name'],var_name='subject',value_name='score')
spt = pd.pivot_table(rshap, index='name', columns='subject', values='score')
spt
```

```python
import pandas as pd
data = {'name':['ahsan','ahsan','mohsin','ali'],
        'urdu':[10,10,16,20],
         'sst':[15,15,14,16]}
df = pd.DataFrame(data)
rshap = pd.melt(df,id_vars=['name'],var_name='subject',value_name='score')
spt = pd.pivot_table(rshap, index='name', columns='subject', values='score', aggfunc=['mean','sum','count','min','max'])
spt
```

<h1>10.	Apply multiple aggregation functions to grouped data.</h1>

```python
import pandas as pd
data = {
    'name': ['ahsan', 'ahsan', 'mohsin', 'mohsin', 'ali', 'ali'],
    'subject': ['urdu', 'sst', 'urdu', 'sst', 'urdu', 'sst'],
    'score': [10, 15, 16, 14, 20, 16]
}

df = pd.DataFrame(data)
aggregated_data = df.groupby('name')['score'].agg(['sum', 'mean', 'min', 'max', 'count'])

print(aggregated_data)
```

<h1>11.	Use the groupby function to group data and apply custom functions.</h1>

```python
import pandas as pd
data = {
    'name': ['ahsan', 'ahsan', 'mohsin', 'mohsin', 'ali', 'ali'],
    'subject': ['urdu', 'sst', 'urdu', 'sst', 'urdu', 'sst'],
    'score': [10, 15, 16, 14, 20, 16]
}


df = pd.DataFrame(data)

aggregated_data = df.groupby('name')['score'].agg(['sum', 'mean', 'min', 'max', 'count'])
aggregated_data
```
