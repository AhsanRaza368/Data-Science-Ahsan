<h1>Fine tuning</h1>

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report



#Step1: Load dataset
data = pd.read_table("winequality-white[1].csv",sep=";")  # Replace with actual path

data['Quality_check']=data['quality'].apply(lambda q: 1 if q>=7 else 0)
data['Quality_check'].value_counts()
data.drop('quality',axis=1,inplace=True)

#Step2: Split features and target
X = data.drop("Quality_check", axis=1)
y = data["Quality_check"]

#Step3: Feature scaling (optional but helps in general)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#Step4: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=43)

#Step5: Hyperparameter grid for tuning
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

#Step6: Initialize GridSearchCV
grid = GridSearchCV(RandomForestClassifier(random_state=43), param_grid, cv=5, scoring='accuracy', n_jobs=-1)

#Step7: Fit the model
grid.fit(X_train, y_train)

#Step8: Best estimator
best_rf = grid.best_estimator_

#Step9: Predictions
y_pred = best_rf.predict(X_test)

#Step10: Evaluation
print("Best Parameters:", grid.best_params_)
print("Train Accuracy:", grid.best_score_)
print("Test Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
