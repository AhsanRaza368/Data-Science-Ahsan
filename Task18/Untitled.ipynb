<h1>Task 18 ->Introduction to Seaborn (visualization with Seaborn) </h1>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[4,5,6,7,8,9]
sns.scatterplot(x,label="y=2x",markers="o",color="blue")
plt.title("Average Order ETA Over Time")
plt.xlabel("Order DateTime")
plt.ylabel("Average ETA (minutes)")
plt.legend()
plt.show()
```

<h1>Style	Description
"darkgrid"	Default Seaborn style with a dark background grid
"whitegrid"	White background with grid (great for statistical plots)
"dark"	Dark background, no grid
"white"	White background, no grid
"ticks"	White background with ticks on axes</h1>

```python

x=[1,2,3,4,5,6]
y=[4,5,6,7,8,9]
sns.scatterplot(x,label="y=2x",markers="o",color="blue")
sns.set_style("darkgrid")
plt.title("Average Order ETA Over Time")
plt.xlabel("Order DateTime")
plt.ylabel("Average ETA (minutes)")
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[4,5,6,7,8,9]
sns.lineplot(x,label="x=2x",color="red")
sns.set_style("whitegrid")
plt.title("Average Order ETA Over Time")
plt.xlabel("Order DateTime")
plt.ylabel("Average ETA (minutes)")
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[4,5,6,7,8,9]
sns.barplot(x,label="x=2x",color="red")
sns.set_style("ticks")
plt.title("Average Order ETA Over Time")
plt.xlabel("Order DateTime")
plt.ylabel("Average ETA (minutes)")
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[4,5,6,7,8,9]
sns.countplot(x,label="x=2x",color="red")
sns.set_style("whitegrid")
plt.title("Average Order ETA Over Time")
plt.xlabel("Order DateTime")
plt.ylabel("Average ETA (minutes)")
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[4,5,6,7,8,9]
sns.boxplot(x,label="x=2x",color="red")
sns.set_style("darkgrid")
plt.title("Average Order ETA Over Time")
plt.xlabel("Order DateTime")
plt.ylabel("Average ETA (minutes)")
plt.legend()
plt.show()
```

<h1>Palette Name	Description
"deep"	Default Seaborn palette
"muted"	Softer colors
"pastel"	Light pastel shades
"bright"	Bright and vivid colors
"dark"	Darker tones
"colorblind"	Colorblind-friendly color palette</h1>

```python
x=[1,2,3,4,5,6]
y=[4,5,6,7,8,9]
sns.scatterplot(x,label="x=2x",color="red")
sns.set_style("darkgrid")
sns.set_palette("colorblind")
plt.title("Average Order ETA Over Time")
plt.xlabel("Order DateTime")
plt.ylabel("Average ETA (minutes)")
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[4,5,6,7,8,9]
sns.lineplot(x,label="x=2x",color="red")
sns.set_style("darkgrid")
sns.set_palette("pastel")
plt.title("Average Order ETA Over Time")
plt.xlabel("Order DateTime")
plt.ylabel("Average ETA (minutes)")
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[4,5,6,7,8,9]
sns.barplot(x,label="x=2x",color="red")
sns.set_style("darkgrid")
sns.set_palette("pastel")
plt.title("Average Order ETA Over Time")
plt.xlabel("Order DateTime")
plt.ylabel("Average ETA (minutes)")
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[4,5,6,7,8,9]
sns.countplot(x,label="x=2x",color="red")
sns.set_style("darkgrid")
sns.set_palette("pastel")
plt.title("Average Order ETA Over Time")
plt.xlabel("Order DateTime")
plt.ylabel("Average ETA (minutes)")
plt.legend()
plt.show()
```
