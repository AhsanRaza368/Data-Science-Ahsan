```python
import seaborn as sns
```

```python


# Visualize outliers in price
sns.boxplot(df['price'])
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['price'] < (Q1 - 1.5 * IQR)) | (df['price'] > (Q3 + 1.5 * IQR)))]
```

```python
max_val=df.max()
max_val
```

```python
min_val=df.min()
min_val
```
