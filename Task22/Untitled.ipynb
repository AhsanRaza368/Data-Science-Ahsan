<h1>Linear regression from scratch</h1>

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
X = 2 * np.random.rand(100, 1)
true_w = 3.5
true_b = 1.2
y = true_w * X + true_b + np.random.randn(100, 1) * 0.5

w = 0.0
b = 0.0


lr = 0.05
epochs = 1000
m = len(X)

losses = []

for epoch in range(epochs):

    y_pred = w * X + b
    

    loss = np.mean((y - y_pred) ** 2)
    losses.append(loss)


    dw = (-2 / m) * np.sum((y - y_pred) * X)
    db = (-2 / m) * np.sum(y - y_pred)
    

    w -= lr * dw
    b -= lr * db

    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Loss = {loss:.4f}, w = {w:.4f}, b = {b:.4f}")


print("\nTraining complete:")
print(f"Final weight (w): {w}")
print(f"Final bias (b): {b}")


plt.scatter(X, y, label="Original data", alpha=0.6)
plt.plot(X, w * X + b, color='red', label="Learned line")
plt.title("Linear Regression from Scratch")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
