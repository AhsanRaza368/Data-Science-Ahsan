<h1>Task 19 ->Statistical plotting with Seaborn (regression plots, distribution plots)</h1>

<h1>1.	Linear Regression Plot (sns.regplot)</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
x=[1,3,5,7,9]
y=[2,4,5,6,7]
sns.regplot(x=x,y=y,label="reg_plot",color="blue")
sns.set_palette("deep")
sns.set_style("darkgrid")
plt.title("my first regression plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

<h1>2.	Multiple Linear Regression Plot (sns.lmplot)</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
x=[1,3,5,7,9]
y1=[0,2,4,5,8]
y2=[3,5,7,9,11]
y3=[2,6,10,14,18]
sns.regplot(x=x,y=y1,label="y=x-1",color="blue")
sns.regplot(x=x,y=y2,label="y=x+2",color="green")
sns.regplot(x=x,y=y3,label="y=2x",color="skyblue")
sns.set_palette("deep")
sns.set_style("darkgrid")
plt.title("my first regression plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

<h1>3.	Residual Plot (sns.residplot)</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
x=[1,3,5,7,9]
y=[0,2,4,5,8]
# y2=[3,5,7,9,11]
# y3=[2,6,10,14,18]
sns.residplot(x=x,y=y,label="y=2x",color="skyblue")
sns.set_palette("deep")
sns.set_style("darkgrid")
plt.title("my first Residual Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

<h1>6.	Pair Plot (sns.pairplot)</h1>

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({
    "x": [1, 3, 5, 7, 9],
    "y": [0, 2, 4, 5, 8]
})

# Styling
sns.set_style("darkgrid")
sns.set_palette("deep")

# Pairplot
sns.pairplot(df)

plt.suptitle("Pairplot of x and y", y=1.02)
plt.show()
```

<h1>1.	Heatmap (sns.heatmap)</h1>

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = {
    "A": [1, 2, 3, 4, 5],
    "B": [5, 4, 3, 2, 1],
    "C": [2, 3, 4, 5, 6]
}

df = pd.DataFrame(data)

corr_matrix = df.corr()

sns.set_style("darkgrid")
sns.heatmap(corr_matrix, annot=True, cmap="YlGnBu", linewidths=0.5)

plt.title("Correlation Heatmap")
plt.show()
```
