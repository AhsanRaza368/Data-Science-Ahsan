<h1>9.	Normalize a column using Min-Max scaling.</h1>

```python
import pandas as pd

df = pd.DataFrame({'value': [10-5, 20-5, 30-5, 40-5, 50-5]})

df['normalized_value'] = (df['value'] - df['value'].min()) / (df['value'].max() - df['value'].min())

print(df)
```

<h1>10.Standardize a column (z-score normalization).<h1>

```python
df = pd.DataFrame({'value': [10-5, 20-5, 15,30-5, 40-5, 50-5]})

df['standardized_value'] = (df['value']-df['value'].mean())/df['value'].std()
df
```

<h1>19.	Extract year, month, and day from a datetime column.</h1>

```python
import pandas as pd

df = pd.DataFrame({
    'value': [5, 5, 15, 30-5, 40-5, 50-5, 15],
    'names': ['ahsan', 'ahsan', 'ali', 'nabi', 'shakoor', 'shamus', 'sahil'],
    'Date': ['2025-04-01', '2025/04/02', 'April 3, 2025', '04-04-2025', 
             '2025-04-01', '2025/04/02', 'April 3, 2025']
})

df['date'] = pd.to_datetime(df['Date'], format='mixed')

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day

print(df)
```

<h1>20.	Filter rows based on a date range.</h1>

```python
import pandas as pd
df = pd.DataFrame({
    'value': [5, 5, 15, 30-5, 40-5, 50-5, 15],
    'names': ['ahsan', 'ahsan', 'ali', 'nabi', 'shakoor', 'shamus', 'sahil'],
    'Date': ['2025-04-01', '2025/04/02', 'April 3, 2025', '04-04-2025', 
             '2025-04-01', '2025/04/02', 'April 3, 2025']
})

df['date'] = pd.to_datetime(df['Date'], format='mixed')
start_date = '2025-04-01'
end_date = '2025-04-02'
filtered_df = df[(df['date'] >= start_date) & (df['date'] <= end_date)]

print(filtered_df)





```

<h1>22.	Convert a categorical column to numerical using label encoding.</h1>

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
df = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'fruit': ['apple', 'banana', 'apple', 'orange']
})

le = LabelEncoder()
df['fruit_encoded'] = le.fit_transform(df['fruit'])

print(df)
```

<h1>23.	Group values in a categorical column and create a new column with grouped categories.</h1>

```python
import pandas as pd
df = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'fruit': ['apple', 'banana', 'apple', 'orange']
})

group_mapping = {
    'apple': 'Group 1',
    'banana': 'Group 1',
    'orange': 'Group 2'
}
df['Grouping'] = df['fruit'].map(group_mapping)

print(df)
```

<h1>28.	Discretize a continuous column into bins.</h1>

```python
import pandas as pd

df = pd.DataFrame({
    'rate': [45, 67, 89, 34, 78, 90, 56]
})

df['rating_group'] = pd.cut(df['rate'], bins=3, labels=['Low', 'Medium', 'High'])

print(df)
```

<h1>29.	Create polynomial features from existing numerical columns.</h1>

```python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
data = {'A': [1, 2, 3, 4], 'B': [5, 6, 7, 8]}
df = pd.DataFrame(data)
poly = PolynomialFeatures(degree=2)
poly_features = poly.fit_transform(df)
poly_df = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(df.columns))

print(poly_df)
```
