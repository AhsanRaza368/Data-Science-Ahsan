```python
df['enrollment_count'] = pd.to_numeric(df['enrollment_count'], errors='coerce')
df['pre_score'] = pd.to_numeric(df['pre_score'], errors='coerce')
df['post_score'] = pd.to_numeric(df['post_score'], errors='coerce')
```

```python
df["pre_score"] = pd.to_numeric(df["pre_score"],errors="coerce")
df["score_dif"]=df["post_score"]-df["pre_score"]
```

```python
df
```

```python
score_improve = df.groupby('course_type')['score_dif'].mean()

plt.bar(score_improve.index, score_improve.values, color=['green', 'purple'])
plt.title("Average Score Improvement by Course Type")
plt.xlabel("Course Type")
plt.ylabel("Score Difference (Post - Pre)")
plt.grid(axis='y')
plt.show()
```

```python
dept_group = df.groupby(['department', 'course_type'])['enrollment_count'].mean().unstack()

dept_group.plot(kind='bar', figsize=(10, 6), color=['lightcoral', 'lightseagreen'])
plt.title("Average Enrollment by Department and Course Type")
plt.xlabel("Department")
plt.ylabel("Average Enrollment")
plt.legend(title='Course Type')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
